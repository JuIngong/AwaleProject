## Test Q&A using a model from HuggingFace

Models : https://huggingface.co/models

Load token from Hugging Face (saved in creds.yml file)

```python
from yaml.loader import SafeLoader
import yaml
# Configuration
with open("C:\Users\JuIng\Documents\Formation\llm/creds.yml") as f:
    config = yaml.load(f, Loader=SafeLoader)

api_token_hf = config['hf_token_2025']
```

```python
model_name = "KingNish/Qwen2.5-0.5b-Test-ft"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
prompt = "Which is greater 9.9 or 9.11 ??"
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```

## Text generation with GPT2

```python
from langchain_huggingface import HuggingFacePipeline, ChatHuggingFace
from langchain_core.prompts import PromptTemplate
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
max_new_tokens = 500
## GPT2 model
hf_gpt = HuggingFacePipeline.from_model_id(
    model_id="gpt2",
    task="text-generation",
    pipeline_kwargs={"max_new_tokens": max_new_tokens},
)
```

```python
## With GPT2
template = """
{question}
"""
prompt = PromptTemplate.from_template(template)
chain = prompt | hf_gpt
question = "Where is Paris?"

print(chain.invoke({"question": question}))
```
